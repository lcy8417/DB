```python
### 실제 활용해보며 익히는 MySQL 기본 + (중급) 문법
# TABLE 분리, FOREIGN KEY, PRIMARY KEY 사용
```

```python
CREATE TABLE ranking (
    num INT AUTO_INCREMENT NOT NULL PRIMARY KEY,
    main_category VARCHAR(50) NOT NULL,
    sub_category VARCHAR(50) NOT NULL,
    item_ranking TINYINT UNSIGNED NOT NULL,
    item_code VARCHAR(20) NOT NULL,
    FOREIGN KEY (item_code) REFFERENCES items(item_code)
);
```

```python
CREATE TABLE items(
    item_code VARCHAR(20) NOT NULL PRIMARY KEY,
    title VARCHAR(200) NOT NULL,
    ori_price INT NOT NULL,
    dis_price INT NOT NULL,
    discount_percent INT NOT NULL,
    provider VARCHAR(100)
);
```

```python
import requests
from bs4 import BeautifulSoup

url = 'http://corners.gmarket.co.kr/Bestsellers'
res = requests.get(url)
soup = BeautifulSoup(res.content, 'html.parser')

categories = soup.select("ul.by-group > li") # 베스트 카테고리

for category in categories:
    get_category('http://corners.gmarket.co.kr' + category.select_one("a")['href'], category.select_one("a").get_text())
```

```python
def get_category(category_link, category_name):
    res = requests.get(category_link)
    soup = BeautifulSoup(res.content, 'html.parser')
    print(category_name)
    sub_categories = soup.select("div.navi.group ul li > a") # 서브 카테고리

    for subcategory in sub_categories:
#         print(subcategory.get_text())
        get_items('http://corners.gmarket.co.kr' + subcategory['href'], category_name, subcategory.get_text())
```

```python
def get_items(html, category_name, subcategory_name):  
    res = requests.get(html)
    soup = BeautifulSoup(res.content, 'html.parser')
    best_list = soup.select('div.best-list')[1]
        
    for ranking, item in enumerate(best_list.select("ul > li"), 1):
        itemname = item.select_one("a.itemname")
        o_price = item.select_one("div.o-price")
        s_price = item.select_one("div.s-price > strong > span")
        dis_percent = item.select_one("div.s-price > span > em")
        
        itemlink = item.select_one("a")['href']
        seller_res = requests.get(itemlink)
        seller_soup = BeautifulSoup(seller_res.content, 'html.parser')
        seller_info = seller_soup.select_one('span.text__brand > span')
        item_code = itemlink.split('=')[1].split('&')[0]
        
        if seller_info == None or seller_info == '':
            seller_info = ''
        else:
            seller_info = seller_info.get_text()
            
        if s_price == None or s_price == '':
            s_price = 0
        else:
            s_price = s_price.get_text().replace(',','').replace('원','')
            
        if o_price == None or o_price == '':
            o_price = s_price
        else:
            o_price = o_price.get_text().replace(',','').replace('원','')
            
        if dis_percent == None or dis_percent == '':
            dis_percent = 0
        else:
            dis_percent = dis_percent.get_text().replace('%','')
            
        print(ranking, category_name, subcategory_name, itemname.get_text(), o_price, s_price, dis_percent, seller_info, item_code)
```
