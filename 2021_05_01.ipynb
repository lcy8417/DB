```python
import requests
from bs4 import BeautifulSoup
import pymysql
db = pymysql.connect(host='localhost', port=3306, user='root', passwd='6818', db='ex', charset='utf8')
cursor = db.cursor()

url = 'http://corners.gmarket.co.kr/Bestsellers'
res = requests.get(url)
soup = BeautifulSoup(res.content, 'html.parser')

best_category = soup.select('#categoryTabG > li > a')

for item in best_category:
    get_category('http://corners.gmarket.co.kr' + item['href'], item.get_text())

db.commit()
db.close()
```

```python
def get_category(category_link, category_name):
    res = requests.get(category_link)
    soup = BeautifulSoup(res.content, 'html.parser')
    sub_categories = soup.select("div.navi.group > ul > li > a")
    for sub_item in sub_categories:
        get_item('http://corners.gmarket.co.kr' + sub_item['href'], category_name , sub_item.get_text())
```

```python
def get_item(category_link, sub_category_name, category_name,):
    
    res = requests.get(category_link)
    soup = BeautifulSoup(res.content, 'html.parser')
    best_items = soup.select('div.best-list')[1].select('ul > li')
    for ranking, best_item in enumerate(best_items, 1):
        dict_item = dict()
        title = best_item.select_one("a.itemname").get_text()
        o_price = best_item.select_one('div.o-price > span')
        s_price = best_item.select_one('div.s-price > strong > span')
        dis_percent = best_item.select_one('div.s-price em')
        
        item_link = best_item.select_one("a")['href']
        item_res = requests.get(item_link)
        item_soup = BeautifulSoup(item_res.content, 'html.parser')
        provider = item_soup.select_one("span.text__seller > a").get_text()
        
        item_code = item_link.split('=')[1].split('&')[0]
        
        if s_price == None or s_price == '':
            s_price = '0'
        else:
            s_price = s_price.get_text().replace(',', '').replace('원','')
            
        if o_price == None or o_price == '':
            o_price = s_price
        else:
            o_price = o_price.get_text().replace(',', '').replace('원','')
            
        if dis_percent == None or dis_percent == '':
            dis_percent = '0';
        else:
            dis_percent = dis_percent.get_text().replace('%','')
        
        dict_item['category_name'] = category_name
        dict_item['sub_category_name'] = sub_category_name
        dict_item['title'] = title
        dict_item['o_price'] = o_price
        dict_item['s_price'] = s_price
        dict_item['dis_percent'] = dis_percent
        dict_item['provider'] = provider
        dict_item['item_code'] = item_code
        dict_item['ranking'] = ranking
        save_items(dict_item)
#         print(dict_item)
#         print(category_name, sub_category_name, ranking, title, o_price, s_price, dis_percent, provider, item_code)
```

```python
import pymysql
db = pymysql.connect(host='localhost', port=3306, user='root', passwd='6818', db='ex', charset='utf8')
cursor = db.cursor()
sql = """CREATE TABLE items(
    item_code VARCHAR(30) NOT NULL PRIMARY KEY,
    title VARCHAR(200) NOT NULL,
    ori_price INT NOT NULL,
    dis_price INT NOT NULL,
    discount_percent INT NOT NULL,
    provider VARCHAR(100)
    );"""
cursor.execute(sql)
db.commit()
sql = """ CREATE TABLE ranking(
    num INT AUTO_INCREMENT NOT NULL PRIMARY KEY,
    main_category VARCHAR(30) NOT NULL,
    sub_category VARCHAR(30) NOT NULL,
    item_ranking TINYINT NOT NULL,
    item_code VARCHAR(10) NOT NULL,
    FOREIGN KEY(item_code) REFERENCES items(item_code)
    );"""
cursor.execute(sql)
db.commit()
db.close()
```

```python
def save_items(dict_item):
    print("ok")
    sql = """SELECT COUNT(*) FROM items WHERE item_code = '""" + dict_item['item_code'] + """';"""
    cursor.execute(sql)
    result = cursor.fetchone()
    if result[0] == 0:
        sql = """INSERT INTO items VALUES('""" + dict_item['item_code'] + """',
        '""" + dict_item['title'] + """',
        """ + dict_item['o_price'] + """,
        """ + dict_item['s_price'] + """,
        """ + dict_item['dis_percent'] + """,
        '""" + dict_item['provider'] + """'
        );
        """
        cursor.execute(sql)
        
    sql = """INSERT INTO ranking(main_category, sub_category, item_ranking, item_code) VALUES('""" + dict_item['category_name'] + """',
    '""" + dict_item['sub_category_name'] + """',
    """ + str(dict_item['ranking']) + """,
    '""" + dict_item['item_code'] + """'
    );
    """
    cursor.execute(sql)
```
